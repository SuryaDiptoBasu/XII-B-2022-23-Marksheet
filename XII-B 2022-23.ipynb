```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
%matplotlib inline
```

```python
data1=pd.read_excel('SECRET.xlsx')
```

```python
data1.head()
```

```python
d={'Name':data1["Student's name"],'English':data1['ENGLISH THEORY']+data1['ENGLISH PRACTICAL'],
  'Hindi/Maths':data1['HINDI/MATH THEORY']+data1['HINDI/MATH PRACTICAL'],'Physics':data1['PHYSICS THEORY']
   +data1['PHYSICS PRACTICAL'],'Chemistry':data1['CHEMISTRY THEORY']+data1['CHEMISTRY PRACTICAL'],
  'Biology':data1['BIOLOGY THEORY']+data1['BIOLOGY PRACTICAL']}
marks=pd.DataFrame(d)
marks['Percentage']=(marks['English']+marks['Hindi/Maths']+marks['Physics']+marks['Chemistry']+marks['Biology'])*100/500
```

```python
marks
```
